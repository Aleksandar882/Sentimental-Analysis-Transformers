```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(df['tweet'],df['sentiment'],test_size=0.2)
```

```python
df_train = pd.concat([X_train,Y_train],axis=1)
```

```python
df_test= pd.concat([X_test,Y_test],axis=1)
```

```python
df_train.columns = ['tweet','label']
df_test.columns = ['tweet','label']
```

```python
df_train.to_csv('train.csv',index=False)
df_test.to_csv('test.csv',index=False)
```

```python
from datasets import load_dataset
```

```python
dataset = load_dataset("csv", data_files={"train": "train.csv", "test": "test.csv"})
```

```python
encoded_dataset = dataset.map(lambda t: tokenizer(t['tweet'],  truncation=True), batched=True,load_from_cache_file=False)
```

```python
model = AutoModelForSequenceClassification.from_pretrained('bert-base-uncased',num_labels =4)
```

```python
encoded_dataset
```

```python
from transformers import TrainingArguments, Trainer
```

```python
arg = TrainingArguments(
    "sentiment",
    learning_rate=5e-5,
    num_train_epochs=1,
    per_device_eval_batch_size=8,
    per_device_train_batch_size=8,
    seed=19
)
```

```python
trainer = Trainer(
    model=model,
    args=arg,
    tokenizer=tokenizer,
    train_dataset=encoded_dataset['train'],
    eval_dataset=encoded_dataset['test']
)
```

```python
trainer.train()
```

```python
y_pred = trainer.predict(encoded_dataset['test'])
```

```python
y_pred
```

```python
y_pred = y_pred.predictions
```

```python
import numpy as np
```

```python
y_pred = [np.argmax(y_pred[i]) for i in range(0,len(y_pred))]
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,f1_score
```

```python
f1_score(df_test['label'],y_pred, average='micro')
```

```python
print(classification_report(df_test['label'],y_pred))
```
